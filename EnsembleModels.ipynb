```python

import pandas as pd

effnet_pred = pd.read_csv('/content/drive/My Drive/Model_predictions/effnet_b3_submission.csv')
resnet_pred = pd.read_csv('/content/drive/My Drive/Model_predictions/resnet_submission.csv')
```

```python
effnet_pred
```

```python
resnet_pred
```

```python
const_df = resnet_pred[['timestamp', 'track_id']]
const_df
```

```python
weight1 = 0.80
weight2 = 0.20
effnet_pred = effnet_pred.multiply(weight1)
resnet_pred = resnet_pred.multiply(weight2)

```

```python
effnet_pred
```

```python
ensemble_resnet_effnet = pd.concat([effnet_pred,resnet_pred]).groupby(level=0).mean()

ensemble_resnet_effnet = ensemble_resnet_effnet.drop(['timestamp', 'track_id'], axis = 1)
ensemble_resnet_effnet
#ensemble_resnet_effnet.to_csv('/content/drive/My Drive/Model_predictions/ensemble_partialdata-resnet_effnet.csv', encoding='utf-8', index=False)
```

```python
ensemble_resnet_effnet = pd.concat([const_df, ensemble_resnet_effnet], axis=1)
```

```python
ensemble_resnet_effnet
```

```python
ensemble_resnet_effnet.to_csv('/content/drive/My Drive/Model_predictions/ensemble_weighted_resnet_effnet_b3.csv', encoding='utf-8', index=False)
```
